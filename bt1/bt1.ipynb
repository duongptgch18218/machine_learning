```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_frame = pd.read_csv('insurance.csv')
x = data_frame.values[:, [0]]
y = data_frame.values[:, [1]]
print(x.shape)
print(y.shape)
```

```python
plt.scatter(y, x, color = 'black')
plt.xlabel('total')
plt.ylabel('Claim')
```

```python
init_range = 0.01
weight = np.random.uniform(low = -init_range, high = init_range, size = (1,1))
bias = np.random.uniform(low = -init_range, high = init_range, size = (1,1))
print(weight)
print(bias)
```

```python

for i in range(100):
    output = np.dot(x, weight) + bias
    delta = output - y
    learning_rate = 0.02
    loss = np.sum(delta**2)/2/63
    print(loss)
    detail_delta = delta/63
    weight = weight - learning_rate*np.dot(x.T,detail_delta)
    bias = bias - learning_rate*np.sum(detail_delta)

print(weight)
print(bias)
```

```python
plt.plot(output, y)
plt.plot(y, x, color = 'black')
plt.show()
```
